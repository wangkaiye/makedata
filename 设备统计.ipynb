```python
import random as r
import pymysql 
import datetime
# 给数据库一次插入多条数据
# 创建连接
con = pymysql.connect(host='192.168.50.92', user='root', password='htdata1234@CD2', database='auto-knowledge-app', port=8859)
# 创建游标
cur = con.cursor()

for i in range(10000): 
    details_list=[]
    
    total_device_number=r.randint(80,220)
    details_list.append(total_device_number)
    power_on_number=int(total_device_number*r.uniform(0.75,0.99))
    details_list.append(power_on_number)
    power_on_rate=power_on_number/total_device_number
    details_list.append(power_on_rate)
    operation_rate=power_on_number/total_device_number-0.05
    details_list.append(operation_rate)
    failure_rate=r.uniform(0.01,0.03)
    details_list.append(failure_rate)
    standby_rate=r.uniform(0.01,0.1)
    details_list.append(standby_rate)
    statistical_time=(datetime.datetime.now()+datetime.timedelta(hours=i)).strftime("%Y-%m-%d %H:%M:%S")
    details_list.append(statistical_time)
    print(details_list)
    sql = '''
    insert into device_statistic(total_device_number, power_on_number, power_on_rate, operation_rate, failure_rate, standby_rate,statistical_time) values(%s, %s, %s, %s, %s, %s, %s)
    '''
    args = details_list
    try:
        # 执行sql
        cur.execute(sql,args) # 插入单条数据，就是单个元组组成的list
       # 执行没问题，就提交事务
        con.commit()
        print('插入数据成功！！')
    except Exception as e:
        print(e)
        con.rollback()
        print('插入数据失败！')
   
        
    
    





```
